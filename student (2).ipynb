```python
# Defining function to remove outliers from the subset
def remove_outliers(df_subset):
    '''removes entries with z-score above 3 for specific columns'''
    variables = ['price','bedrooms', 'bathrooms', 'sqft_living', 'sqft_above', 'sqft_living15','floors','numerical_grade']
    
    for variable in variables:
        df_subset= df_subset[np.abs(df_subset[variable]-df_subset[variable].mean()) <= (3*df_subset[variable].std())]
        
    return df_subset
#Removing the outliers which have a standard deviation greater than 3
df_subset = remove_outliers(df_subset)
```

### Model 1

#### Simple linear regression

For our baseline model we chose square foot living as the predictor variable because it has the highst correlation at 0.71 and it's scatter plot against price indicates a potential linear relationship.

```python
y = df_subset["price"]
X_baseline = df_subset[["sqft_living"]]
```

```python
baseline_model = sm.OLS(y, sm.add_constant(X_baseline))
baseline_results = baseline_model.fit()

print(baseline_results.summary())
```

* The model is statistically significant overall, with an F-statistic p-value well below 0.05
* The model explains about 37% of the variance 
* The model coefficients (price and sqft_living) are both statistically significant, with t-statistic p-values well below 0.05
* The coefficient represents the estimated effect of the features on the price. For example the coefficient  for "sqft_living" is 218.0400, indicating that a one-unit increase in square footage of living space is associated with an estimated increase of 0.5332 in the price.

```python
#Plotting the fit of the baseline model on the "sqft_living" feature
sm.graphics.plot_fit(baseline_results, "sqft_living")
plt.show()
```

```python
# Line of best fit re-plotted
fig, ax = plt.subplots()
df_subset.plot.scatter(x="sqft_living", y="price", label="Data points", ax=ax)
sm.graphics.abline_plot(model_results=baseline_results, label="Regression line", ax=ax, color="black")
ax.legend();
```

The regression line on the scatter plot indicates there is a somewhat linear relationship between price and sqft_living but not perfectly linear.

```python
# Calculating Residuals for further Analysis
# Residuals are the difference between  observed values and predicted values of the dependent variable 
resids_1= baseline_results.resid
```

```python
# Checking residuals to assess homoscedasticity
fig, ax = plt.subplots()
ax.scatter(df_subset["price"], resids_1)
ax.axhline(y=0,color = "black")
ax.set_xlabel("sqrt frt living")
ax.set_ylabel("residuals")
ax.set_title("Residuals vs. sqrt frt living(Homoscedasticity Check)");
```

For our baseline model residuals are not evenly spread further suggesting the model is not adequately capturing the variation in the data, hence the need to further improve the model.

```python
# A QQ plot to see if an S curve, therefore forms normal distribution
plt.figure(figsize=(4,4))
fig = sm.graphics.qqplot(resids_1, dist=stats.norm, line='45', fit=True)
plt.title("QQ Plot of Residuals (Normality Check)")
fig.show();
```

The point deviation from the line towards the lower and  upper end suggests skewness in our data.

Generally, the  plot visually depicts extreme deviations from the straight line, indicating non-normality in the data hence the need to further improve our model.

#####  Interpretation
* R-squared: The R-squared value of 0.808 indicates that approximately 80% of the variance in the dependent variable (price) can be explained by the independent variables included in the model.
* Model significance: The F-statistic of  692.2 with a probability (p-value) of 0.00 suggests that the overall model is statistically significant, meaning that at least one of the independent variables has a significant effect on the dependent variable.
* The coefficient represents the estimated effect of the features on the price. 
* The intercept coefficient is 1.441e+05  (144,100), indicating the estimated price when all other independent variables are zero.The coefficient for sqft_living is 112.0845, suggesting that a one-unit increase in square footage of living space is associated with an estimated increase in price of 112.0845 units, holding other variables constant.
* Similarly, other independent variables such as sqft_above, sqft_living15, waterfront status, number of bedrooms and bathrooms, zip code, and so on, have coefficients that represent their estimated impact on the price.
